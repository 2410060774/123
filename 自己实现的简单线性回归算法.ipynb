```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5],dtype=np.float)
y=np.array([1,3.0,2,3,5])
y.dtype
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
num=0.0
d=0.0
```

```python
for x_i,y_i in zip(x,y):
    num+=(x_i-x_mean)*(y_i-y_mean)
    d+=(x_i-x_mean)**2
    a=num/d
    b=y_mean-a*x_mean
y_hat=a*x+b    
```

```python
plt.scatter(x,y);
plt.plot(x,y_hat,c='r')
plt.grid()
```

```python
x_predict=4.6
y_predict=a*x_predict+b
y_predict
```

```python
plt.scatter(x_predict,y_predict,c='cyan',marker='*')
plt.scatter(x,y);
plt.plot(x,y_hat,c='r');
plt.grid();
```
